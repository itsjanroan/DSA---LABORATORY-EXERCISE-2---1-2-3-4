```python
#CREATOR INFORMATION
#Student Name: Abad, Jan Roan D.
#Section: BSCpE 2-2
#Task: Lab 2 Exercise 1 - A teacher named Mrs. Rivera who loved making math fun for her students. One day, a student named Mia asked how to calculate powers, like (2^7). Mrs. Rivera explained, "Imagine you need to multiply 2 by itself seven times. We can solve this sing a python code utilizing special method called recursion to solve this," Can you help the class of Mrs. Rivera by providing the recursive method to calculate powers?
# This function calculates powers using recursion
def calculate_power(base, exponent):
    # If the exponent is negative, we flip it and take the reciprocal of the result
    if exponent < 0:
        return 1 / calculate_power(base, -exponent)
    
    # Base case: If the exponent is 0, return 1 (since anything to the power of 0 is 1)
    if exponent == 0:
        return 1
    
    # Otherwise, keep multiplying the base by itself until we reach the base case
    return base * calculate_power(base, exponent - 1)

# This loop will keep the program running until the user decides to stop
while True:
    # Let's get the base and exponent from the user
    base_number = float(input("Enter the base number: "))
    exponent_number = int(input("Enter the exponent: "))

    # Now we calculate the result using our recursive function
    result = calculate_power(base_number, exponent_number)

    # Show the result in a simple, readable format
    print(f"{base_number} raised to the power of {exponent_number} is: {result}")

    # Check if the user wants to try another calculation
    try_again = input("Do you want to try again? (yes/no): ").lower()

    # If they say 'no', we'll break out of the loop and say goodbye
    if try_again == 'no':
        print("Goodbye!")
        break
```

```python
#CREATOR INFORMATION
#Student Name: Abad, Jan Roan D.
#Section: BSCpE 2-2
#Task: Lab 2 Exercise 3 - The code engages users by asking them for the length of the sides of a square they wish to create. It allows them to input their preferences and provides immediate visual feedback based on their input.
# Function to create a hollow square
# Let's create a hollow square together!
def generate_hollow_square(side_length):
    # We're going to loop through each row of the square
    for row in range(side_length):
        # Now, let’s look at each column in that row
        for col in range(side_length):
            # If we're on the border of the square, we'll print 'x'
            if row == 0 or row == side_length - 1 or col == 0 or col == side_length - 1:
                print('x', end=' ')
            else:
                # Otherwise, we'll just leave a space for the hollow part
                print(' ', end=' ')
        # After finishing the row, we'll move to the next line
        print()

# Here we go! Let's ask the user for their input
while True:
    try:
        length = int(input("Enter the side length of the square: "))
        if length < 1:
            print("Oops! Please enter a positive integer greater than zero.")
        else:
            generate_hollow_square(length)
    except ValueError:
        print("Hmm, that doesn't seem like a valid number! Please try again with a positive integer.")
    
    # Asking the user if they want to keep going
    continue_choice = input("Do you want to create another hollow square? (yes/no): ").strip().lower()
    if continue_choice != 'yes':
        print("Thanks for playing! Goodbye!")
        break  # Exit the loop and close the program
```

```python
#CREATOR INFORMATION
#Student Name: Abad, Jan Roan D.
#Section: BSCpE 2-2
#Task: Lab 2 Exercise 4 - The code engages users by asking them for the length of the sides of a square they wish to create. It allows them to input their preferences and provides immediate visual feedback based on their input.
# Let’s create an inverted right triangle together!

def draw_inverted_triangle(height):
    # Looping from the height down to 1 to create the inverted triangle
    for i in range(height, 0, -1):
        # Printing a certain number of stars for each row
        print('* ' * i)

# Time to get started! Let’s ask the user for the height of the triangle
while True:
    try:
        # Asking for the height
        n = int(input("Please enter the height of the triangle: "))
        if n < 1:
            print("Oops! That doesn't seem right. Please enter a positive integer.")
        else:
            draw_inverted_triangle(n)
    except ValueError:
        print("Hmm, that doesn't look like a valid number. Can you try again with a positive integer?")
    
    # Now, let's see if they want to draw another triangle
    while True:
        repeat = input("Would you like to create another inverted triangle? (yes/no): ").strip().lower()
        if repeat == 'yes':
            break  # Breaks out of the inner loop to ask for the height again
        elif repeat == 'no':
            print("Thanks for joining in! Goodbye!")
            exit()  # Exits the program
        else:
            print("Please respond with 'yes' or 'no'.")  # Prompt for valid response
```
